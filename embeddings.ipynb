```python
import pandas as pd
import numpy as np
```

```python
train_data = pd.read_csv('/content/train_preprocessed.csv')
```

```python
train_data.head(10)
```

```python
from sentence_transformers import SentenceTransformer
model = SentenceTransformer('paraphrase-MiniLM-L6-v2')
```

```python
encoded_reviews = []
for idx, review in enumerate(train_data['review']):
  encoded_reviews.append(model.encode(review))
```

```python
train_arr = np.array(encoded_reviews)
```

```python
train_arr
```

```python
np.save('train.npy', train_arr)
```
